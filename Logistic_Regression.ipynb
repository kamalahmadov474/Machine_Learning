1.Import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

2.Generate some data points

```python
np.random.seed(0)
X = np.linspace(-10, 10 , 100)
y = (X > 0).astype(int)
```

3.Fit a logistic regression model

```python
model = LogisticRegression()
model.fit(X.reshape(-1,1), y)
```

4.Generate probabilities using the logistic function

```python
X_test = np.linspace(-10, 10 ,300)
y_prob = model.predict_proba(X_test.reshape(-1,1))[:,1]
```

5.Plotting the graph

```python
plt.figure(figsize = (10,6))
plt.scatter(X,y,color = 'blue', label = 'Data points')
plt.plot(X_test,y_prob,color='red',label = 'Logistic Regression (Sigmoid Curve)')
plt.axhline(0.5, color = 'gray', linestyle = '--', label = 'Decision Boundary (y = 0.5)')
plt.axvline(0,color = 'green',linestyle = '--',label = 'Threshold (X = 0)')
plt.xlabel('Feature')
plt.ylabel('Probability of Outcome = 1')
plt.title('Logistic Regression: Graphical Representation')
plt.legend()
plt.grid(True)
plt.show()
```
